```python

import pandas as pd
pd.set_option('display.max_columns', 999)

import matplotlib
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

location = 'C:/Users/jyc96/OneDrive/Documents/Datafest2017/'
dest = pd.read_csv(location + "dest.txt", sep='\t')
data = pd.read_csv(location + "data.txt", sep='\t')
```

```python
visit_freq = data.groupby('user_id')['user_id'].count()
for i in [1, 2, 10, 50, 100, 200]:
    print('users with less than {} sessions: {}%'.format(i, round((visit_freq < i).mean() * 100, 2)))

plt.figure(figsize=(12, 6))
plt.hist(visit_freq.values, bins=50, log=True)
plt.xlabel('Number of sessions', fontsize=12)
plt.ylabel('log(number of users)', fontsize=12)
plt.show()
```

```python
dates = data[['date_time','user_id','srch_ci','srch_co']]
dates['srch_ci']=pd.to_datetime(dates['srch_ci'], errors='coerce')
pd.options.mode.chained_assignment = None
dates['srch_co']=pd.to_datetime(dates['srch_co'], errors='coerce')
dates['duration'] = dates['srch_co']- dates['srch_ci']
```

```python
conversion = data.groupby('user_id')['is_booking'].sum()
```

```python
visit_freq = pd.DataFrame({'visit_freq' : visit_freq}).reset_index()
bookings = pd.DataFrame({'bookings' : data.groupby('user_id')['is_booking'].sum()}).reset_index()
```

```python
visit_bookings = visit_freq.join(bookings.set_index('user_id'), on='user_id', how='right')
```

```python
visit_bookings['conversion'] = visit_bookings['bookings']/visit_bookings['visit_freq']
```

```python
for i in [0 ,5 ,10, 20, 30, 40, 50, 80, 100]:
    print('users with at most {}% conversion: {}%'.format(i, round((visit_bookings['conversion']*100 <= i).mean() * 100, 2)))

plt.figure(figsize=(12, 6))
plt.hist(conversion, bins=10, log=True)
plt.xlabel('conversion', fontsize=12)
plt.ylabel('log(number of users)', fontsize=12)
plt.show()
```
